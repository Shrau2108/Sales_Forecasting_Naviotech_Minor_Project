```python
df = pd.read_csv("train.csv")

# Convert date columns properly (DD/MM/YYYY format)
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True, format='mixed')
df["Ship Date"]  = pd.to_datetime(df["Ship Date"],  dayfirst=True, format='mixed')

df = df.sort_values("Order Date")
df.head()
```

```python
# Shape of dataset
print("Dataset Shape:", df.shape)

# Info â€” datatypes + null count
df.info()

# Missing values column-wise
print("\nMissing Values:\n", df.isna().sum())

# Distribution summary for target variable
print("\nSales Value Summary:\n", df["Sales"].describe())
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month

plt.figure(figsize=(12,4))
sns.lineplot(data=df.groupby(["Year","Month"])["Sales"].sum().reset_index(),
             x="Month", y="Sales", hue="Year", marker="o")
plt.title("Month-wise Sales Comparison per Year")
plt.show()
```

```python
# 1) Remove columns that have no predictive value for sales forecasting
df = df.drop(columns=["Row ID", "Customer Name", "Product Name"])

# 2) Handle missing Postal Code
# Only 11 missing â†’ impact is negligible for forecasting
df["Postal Code"] = df["Postal Code"].fillna(-1)   # or df.dropna(subset=["Postal Code"])
# We assign -1 to represent "unknown code"

# 3) Check duplicate records
duplicate_count = df.duplicated().sum()
print("Duplicate Rows Found:", duplicate_count)

# (Optional) Remove duplicates if necessary
if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates Removed")

# 4) Convert categorical columns to 'category' dtype for memory efficiency
cat_cols = ["Ship Mode", "Segment", "Country", "City", "State", "Region",
            "Product ID", "Category", "Sub-Category", "Customer ID", "Order ID"]

for col in cat_cols:
    df[col] = df[col].astype("category")

# 5) Final validation
df.info()
```

```python
# Build daily sales time series
ts = df.groupby("Order Date")["Sales"].sum().sort_index()
ts = ts.asfreq("D", fill_value=0)   # ensure consistent daily frequency
ts.head()

```

```python
# Convert Series â†’ ML DataFrame
data = pd.DataFrame({"y": ts})
data["date"] = data.index

# 5.1 Lag Features (memory of previous days)
for lag in [1, 2, 7, 14, 28]:
    data[f"lag_{lag}"] = data["y"].shift(lag)

# 5.2 Rolling Window Features (trend + demand consistency)
data["roll_mean_7"]  = data["y"].rolling(7).mean()
data["roll_mean_30"] = data["y"].rolling(30).mean()
data["roll_std_7"]   = data["y"].rolling(7).std()

# 5.3 Calendar Signals
data["dayofweek"]  = data["date"].dt.dayofweek
data["month"]      = data["date"].dt.month
data["year"]       = data["date"].dt.year
data["is_weekend"] = (data["dayofweek"]>=5).astype(int)

# --- Salary cycle features ---
data["day_of_month"] = data["date"].dt.day

# salary credited at start of month
data["is_salary_week"] = data["day_of_month"].between(1, 7).astype(int)

# month-end effect
data["is_month_end"] = data["day_of_month"].between(25, 31).astype(int)

# distance from salary date (1st)
data["days_from_salary"] = (data["day_of_month"] - 1).abs()

# optional: compress to 0..1 impact near salary
data["salary_impact"] = np.where(
    data["days_from_salary"] <= 7,
    1 / (data["days_from_salary"] + 1),
    0
)


# 5.4 Remove rows with NA due to lags/rolling
data = data.dropna()

data.head()
```

```python
data = data.sort_values("date")  # ensure order

split_date = "2018-07-01"   # last 6 months = testing
train = data[data["date"] < split_date]
test  = data[data["date"] >= split_date]

feature_cols = [
    "lag_1", "lag_2", "lag_7", "lag_14", "lag_28",
    "roll_mean_7", "roll_mean_30", "roll_std_7",
    "dayofweek","month","year","is_weekend",
    "day_of_month","is_salary_week","is_month_end",
    "days_from_salary","salary_impact"
]


X_train, y_train = train[feature_cols], train["y"]
X_test,  y_test  = test[feature_cols],  test["y"]

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

```python
future_steps = 60   # change to 30 / 90 / 180 as needed

future_data = data.copy()

for i in range(future_steps):
    last_row = future_data.iloc[-1].copy()

    new_row = {}

    # Target for prediction
    y_pred_next = xgb.predict(last_row[feature_cols].values.reshape(1,-1))[0]
    new_row["y"] = y_pred_next

    # Create new lag values
    for lag in [1,2,7,14,28]:
        new_row[f"lag_{lag}"] = future_data.iloc[-lag]["y"] if len(future_data) >= lag else y_pred_next

    # Rolling features
    new_row["roll_mean_7"]  = future_data["y"].tail(7).mean()
    new_row["roll_mean_30"] = future_data["y"].tail(30).mean()
    new_row["roll_std_7"]   = future_data["y"].tail(7).std()

    # Calendar progression
    next_date = last_row["date"] + pd.Timedelta(days=1)
    new_row["date"] = next_date
    new_row["dayofweek"] = next_date.dayofweek
    new_row["month"] = next_date.month
    new_row["year"] = next_date.year
    new_row["is_weekend"] = int(next_date.dayofweek >= 5)

    future_data = pd.concat([future_data, pd.DataFrame([new_row])], ignore_index=True)

future_forecast = future_data.tail(future_steps)
future_forecast[["date","y"]].head()
```

```python
!pip install optuna --quiet
import optuna

def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 500, 3500),
        "learning_rate": trial.suggest_float("learning_rate", 0.005, 0.05),
        "max_depth": trial.suggest_int("max_depth", 4, 14),
        "subsample": trial.suggest_float("subsample", 0.7, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.7, 1.0),
        "reg_lambda": trial.suggest_float("reg_lambda", 0, 5),
        "reg_alpha": trial.suggest_float("reg_alpha", 0, 5),
        "tree_method": "hist"
    }

    model = XGBRegressor(**params)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    return mean_absolute_error(y_test, pred)

study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=40)   # Increase trials later for more accuracy

print("\nBest Hyperparameters:\n", study.best_params)
```

```python
# How many days into the future to forecast
future_steps = 90    # change to 30 / 60 / 180 etc.

future = data.copy().reset_index(drop=True)

for i in range(future_steps):

    last_date = future["date"].iloc[-1]
    next_date = last_date + pd.Timedelta(days=1)

    row = {
        "date": next_date,

        # ===== Calendar =====
        "dayofweek": next_date.weekday(),
        "month": next_date.month,
        "year": next_date.year,
        "is_weekend": int(next_date.weekday() >= 5),

        # ===== New Salary Cycle Features ===== ðŸš€
        "day_of_month": next_date.day,
        "is_salary_week": int(next_date.day <= 7),
        "is_month_end": int(next_date.day >= 25),
        "days_from_salary": abs(next_date.day - 1),
        "salary_impact": 1/(abs(next_date.day - 1)+1) if abs(next_date.day - 1) <= 7 else 0,
    }

    # ===== LAGS =====
    for lag in [1,2,7,14,28]:
        row[f"lag_{lag}"] = future["y"].iloc[-lag]

    # ===== ROLLING =====
    row["roll_mean_7"]  = future["y"].tail(7).mean()
    row["roll_mean_30"] = future["y"].tail(30).mean()
    row["roll_std_7"]   = future["y"].tail(7).std()

    # Convert aligned to required model feature list
    X_next = pd.DataFrame([row])[feature_cols]   # feature_cols = all 17 features

    # Predict & append next value
    row["y"] = final_xgb.predict(X_next)[0]
    future = pd.concat([future, pd.DataFrame([row])], ignore_index=True)


# ------------------------
# FINAL FUTURE FORECAST CSV
# ------------------------
future_forecast = future.tail(future_steps)[["date","y"]].rename(columns={"y":"forecast"})
future_forecast.head()
```

```python
import pandas as pd

df = pd.read_csv("./data/feature/balajin_sorted.csv")

# Reverse dataset because it is in descending timestamp
df = df.iloc[::-1].reset_index(drop=True)

# Detect numeric sales column automatically
sales_col = df.select_dtypes(include=["float","int"]).columns[0]

# ==========================================
# 108 Months â†’ Jan 2016 â†’ Dec 2024
# ==========================================
start_date = "2016-01-01"
total_months = 108

monthly_dates = pd.date_range(start=start_date, periods=total_months, freq="ME")

rows_per_month = len(df) // total_months   # â‰ˆ 590 per month

# Assign month index correctly after reversing order
df["month_index"] = df.index // rows_per_month
df["date"] = df["month_index"].apply(lambda i: monthly_dates[i] if i < total_months else monthly_dates[-1])

# ==========================================
# Aggregate sales monthly
# ==========================================
monthly = df.groupby("date")[sales_col].sum().reset_index()
monthly.rename(columns={sales_col:"monthly_sales"}, inplace=True)

monthly.to_csv("balaji_monthly_cleaned8.csv", index=False)

print("âœ” Dataset reversed and monthly time-series generated successfully!")
monthly.head()
```

```python
import pandas as pd
import numpy as np

#=========== CONFIG (edit if needed) ===========
FILE = "cisco.csv"          # change to new file
START_YEAR = 2016           # beginning of sales history
END_YEAR = 2024             # estimated last date in file
#==============================================

# LOAD RAW DATA
df = pd.read_csv(FILE)

#------ 1) Reverse if sorted newestâ†’oldest ------
# Heuristic: If first rows > last rows â†’ reverse
if df.select_dtypes(include=["float","int"]).iloc[:200].mean().mean() < df.select_dtypes(include=["float","int"]).iloc[-200:].mean().mean():
    print("ðŸ”„ Data appears reverse sorted â†’ fixing...")
    df = df.iloc[::-1].reset_index(drop=True)
else:
    print("âœ” Already chronological")

#------ 2) Detect numeric sales column ---------
num_cols = df.select_dtypes(include=["float","int"]).columns
sales_col = num_cols[0]   # first numeric column used
print(f"ðŸ“Œ Sales column detected = {sales_col}")

#------ 3) Compute expected monthly length -----
months = (END_YEAR - START_YEAR + 1) * 12
rows_per_month = len(df)//months
print(f"ðŸ“… Aggregating approx {rows_per_month} rows per month")

#------ 4) Date Range Generator ----------------
dates = pd.date_range(f"{START_YEAR}-01-01", periods=months, freq="M")

df["month_index"] = df.index // rows_per_month
df["date"] = df["month_index"].apply(lambda i: dates[i] if i < len(dates) else dates[-1])

#------ 5) Aggregate to monthly ----------------
monthly = df.groupby("date")[sales_col].sum().reset_index()
monthly.rename(columns={sales_col:"monthly_sales"}, inplace=True)

# SAVE
output = FILE.replace(".csv","_monthly_cleaned.csv")
monthly.to_csv(output, index=False)

print(f"\nðŸŽ‰ Monthly Forecasting Dataset Ready!\nSaved as â†’ {output}\n")
monthly.head(12)
```
